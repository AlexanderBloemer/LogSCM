# Block Heuristic
The task is to place parcels on a pallett in specific blocks of the same orientation.
The question is how many parcels should be in each block to maximize number of parcels but respect pallett size.
<img src="img/BlockHeuristic.png" width="400">

```python
# parameters
A = 80
B = 120
a = 27
b = 18
```

```python
block_sides = {
    "A1": b,
    "A2": b,
    "A3": a,
    "B1": a,
    "B2": a,
    "B3": b
}

max_ = {}
for s in block_sides:
    max_[s] = int((A if 'A' in s else B)/block_sides[s])
print(f"maxima: {max_}")
```

```python
import pandas as pd

patterns = []
for i in range(max_["B1"]+1):
    for j in range(max_["A2"]+1):
        patterns.append({
            "A1": int(A/block_sides["A1"]),
            "A2": j,
            "A3": int((A-j*block_sides["A2"])/block_sides["A3"]),
            "B1": i,
            "B2": int((B-i*block_sides["B1"])/block_sides["B2"]),
            "B3": int((B-i*block_sides["B1"])/block_sides["B3"])
        })

# compute number of parcels for each pattern
for p in patterns:
    p["parcels"] = p["A1"]*p["B1"] + p["A2"]*p["B2"] + p["A3"]*p["B3"]

# order the patterns descending by number of parcels
patterns = sorted(patterns, key=lambda item: item["parcels"], reverse=True)

pd.DataFrame(patterns)
```
